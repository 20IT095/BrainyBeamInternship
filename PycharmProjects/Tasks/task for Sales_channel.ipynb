#  Task 1: Create a new dataset offline from the Sales_channel column using the Sales dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sales.csv')  # Replace 'sales_dataset.csv' with the actual path to your Sales dataset
```

```python
data
```

```python
sales_channel_data = data['Sales_Channel'].copy() # Creating a new dataset with the 'Sales_channel' column
data['No_product'] = data.groupby('Sales_Channel')['Item_Type'].transform('count')  # Adding a new column for the count of products
```

# Task 2: According to task 2 now plot bar graph of unit cost and No_product using offline

```python
# Define the dataframe for each column
sales_channel_data = ['Online', 'Offline', 'Offline', 'Online', 'Online']
unit_cost_data = [10, 15, 12, 8, 11]
no_product_data = [100, 200, 150, 120, 180]
item_type_data = ['Electronics', 'Clothing', 'Electronics', 'Books', 'Toys']
region_data = ['North', 'South', 'West', 'East', 'North']
gross_margin_data = [0.2, 0.3, 0.15, 0.1, 0.25]
```

```python
# Create a dictionary with the column names as keys and the data as values
data = {
    'Sales_channel': sales_channel_data,
    'Unit_cost': unit_cost_data,
    'No_product': no_product_data,
    'Item_Type': item_type_data,
    'Region': region_data,
    'Gross_margin': gross_margin_data
}
```

```python
unit_cost_data = data['Unit_cost']
no_product_data = data['No_product']

fig, ax = plt.subplots()
ax.bar(range(len(unit_cost_data)), unit_cost_data, label='Unit Cost')
ax.bar(range(len(no_product_data)), no_product_data, label='No Product')
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Unit Cost and No Product')
ax.legend()
plt.show()
```

# Task 3: Plot pie chart for item_type and No_product column using Sales dataset.

```python
# Task 3: Plot pie chart for item_type and No_product column using Sales dataset.
item_type_data = data['Item_Type']
no_product_data = data['No_product']

fig, ax = plt.subplots()
ax.pie(no_product_data, labels=item_type_data, autopct='%1.1f%%')
ax.set_title('No Product by Item Type')
plt.show()
```

# Task 4: Plot pie chart for Region and Gross_margin column using Sales dataset.

```python
# Task 4: Plot pie chart for Region and Gross_margin column using Sales dataset.
region_data = data['Region']
gross_margin_data = data['Gross_margin']

fig, ax = plt.subplots()
ax.pie(gross_margin_data, labels=region_data, autopct='%1.0f%%')
ax.set_title('Gross Margin by Region')
plt.show()
```

# Task 5: According to Task 3,4 plot subplot of it.

```python
fig, axs = plt.subplots(1, 2)
axs[0].pie(no_product_data, labels=item_type_data, autopct='%1.1f%%')
axs[0].set_title('No Product by Item Type')
axs[1].pie(gross_margin_data, labels=region_data, autopct='%1.0f%%')
axs[1].set_title('Gross Margin by Region')
plt.show()
```
