# seaborn
#### seaborn is used as data visualization and also gives as hight level interface  for creative information

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df['Age'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
x = pd.pivot_table(df,index='Survived',columns='Sex',values='Age')
```

```python
x
```

```python
sns.heatmap(x,annot=True,cmap='coolwarm')
```

```python
cor=df.corr()
```

```python
cor
```

```python
sns.heatmap(cor,annot=True,cmap='Reds')
```

```python
pt = pd.read_csv('sales.csv')
```

```python
pt
```

```python
pt['Year']=pd.DatetimeIndex(pt['Ship_Date']).year
```

```python
pt['Month']=pd.DatetimeIndex(pt['Ship_Date']).month
```

```python
pt
```

total profit,year,month create heatmap for this 

```python
pt = pd.pivot_table(pt,index='Total_Profit',columns='Year',values='Month')
```

```python
pt
```

```python
sns.heatmap(pt,annot=True,cmap='Reds')
```

```python
correlation=pt.corr()
```

```python
correlation
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```
